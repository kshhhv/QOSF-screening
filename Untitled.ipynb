```python
def param_qc(theta):
    qc = QuantumCircuit(2,2)
    qc.rx(pi,0)
    qc.rx(pi,1)
    qc.rx(theta, 0)
    qc.ry(theta, 0)
    qc.cx(0,1)
    qc.measure(range(2), range(2))
    return qc
```

```python
def cost_function(circuit, shots):
    
    p00, p01, p10, p11 = 0,0,0,0
    
    simulator = Aer.get_backend('qasm_simulator')
    counts = execute(circuit, backend=simulator, shots=shots).result().get_counts(circuit)
    
    if '00' in counts.keys():
        p00 = counts['00']/shots
    if '01' in counts.keys():
        p01 = counts['01']/shots
    if '10' in counts.keys():
        p10 = counts['10']/shots
    if '11' in counts.keys():
        p11 = counts['11']/shots

    cost = (0.5 - p01)**2 + (0.5 - p10)**2 
    
    return cost
```

```python
def opt(shots, params, dx):
    
    params_0 = params + dx
    circuit = param_qc(params_0)
    cost_0 = cost_function(circuit, shots)
    
    
    params_1 = params - dx
    circuit = param_qc(params_1)
    cost_1 = cost_function(circuit, shots)    
    
    del_cost = (cost_0 - cost_1)

    return cost_0, del_cost
```

```python
def baap(shot):
    pi = np.pi

    params = 0
    shots = shot
    costs_x = []
    costs_y = []
    iteration = 500
    step = 0.004

    for i in range(iteration):

        cost, grad = opt(shots, params, 0.01)
        costs_x.append(cost)

        #params = [params[0] - grad*step, params[1] - grad*step]

        if grad <= 0:
            params = params + 0.02
        else:
            params = params - 0.02

    param_x = params
    print(param_x)
    
    return costs_x
```

```python
from qiskit.quantum_info import Statevector

shots = [50, 100, 250, 1000]
costs = []
for shot in shots:
    costs.append(baap(shot))
    
```
